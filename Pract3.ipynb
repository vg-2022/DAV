```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randint(0,100,size=(50,3)), columns=list('123'))

for c in  df.sample(int(df.shape[0]*df.shape[1]*0.10)).index:
    df.loc[c,str(np.random.randint(1,4))]=np.nan
df
```

```python
#a
print(df.isnull().sum().sum())
```

```python
#b
for col in df.columns:
    if((df[col].isnull().sum())>5):
        df.drop(columns=[col])
df
```

```python
#c
sum=df.sum(axis=1)
print("Sum :",sum)
print("Max sum:",sum.max())
maxSum = df.sum(axis=1).idxmax()
print("Row Label having maximum sum:" )
print(maxSum)

df = df.drop(maxSum ,axis =0)
print("DF after removing row of maximum sum value")
df
```

```python
#d
sortdf=df.sort_values('1')
sortdf
```

```python
#e
df =df.drop_duplicates(subset='1',keep = "first")
df
```

```python
#f
correlation = df['1'].corr(df['2'])
print("Correlation 1&2: ", correlation)
covariance = df['2'].cov(df['3'])
print("Covariance col 2&3 :",covariance)
```

```python
#g
q1= df.quantile(0.25)
q3=df.quantile(0.75)
print(q3, q1)
print(len(df))
IQR = q3 - q1
print(IQR)

upper_bound = q3 + 1.5 * IQR
lower_bound = q1 - 1.5 * IQR
print(upper_bound)
print(lower_bound)

d1 = df[df < upper_bound]
d2= df[df > lower_bound]
print(len(df))
```
